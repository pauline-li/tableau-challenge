```python
import csv
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Combine csv files for January 2019 to June 2020

# 2020 Trip Histories Data
trip_2019_01 = pd.read_csv("Resources/JC-201901-citibike-tripdata.csv")
trip_2019_02 = pd.read_csv("Resources/JC-201902-citibike-tripdata.csv")
trip_2019_03 = pd.read_csv("Resources/JC-201903-citibike-tripdata.csv")
trip_2019_04 = pd.read_csv("Resources/JC-201904-citibike-tripdata.csv")
trip_2019_05 = pd.read_csv("Resources/JC-201905-citibike-tripdata.csv")
trip_2019_06 = pd.read_csv("Resources/JC-201906-citibike-tripdata.csv")
trip_2019_07 = pd.read_csv("Resources/JC-201907-citibike-tripdata.csv")
trip_2019_08 = pd.read_csv("Resources/JC-201908-citibike-tripdata.csv")
trip_2019_09 = pd.read_csv("Resources/JC-201909-citibike-tripdata.csv")
trip_2019_10 = pd.read_csv("Resources/JC-201910-citibike-tripdata.csv")
trip_2019_11 = pd.read_csv("Resources/JC-201911-citibike-tripdata.csv")
trip_2019_12 = pd.read_csv("Resources/JC-201912-citibike-tripdata.csv")
trip_2020_01 = pd.read_csv("Resources/JC-202001-citibike-tripdata.csv")
trip_2020_02 = pd.read_csv("Resources/JC-202002-citibike-tripdata.csv")
trip_2020_03 = pd.read_csv("Resources/JC-202003-citibike-tripdata.csv")
trip_2020_04 = pd.read_csv("Resources/JC-202004-citibike-tripdata.csv")
trip_2020_05 = pd.read_csv("Resources/JC-202005-citibike-tripdata.csv")
trip_2020_06 = pd.read_csv("Resources/JC-202006-citibike-tripdata.csv")

Ride_2019_04_06 = pd.read_csv("Resources/DailyRideMbr-2019-Apr-Jun.csv")



# 2019 Trip Histories Data
# trip_2019_04 = pd.read_csv("Resources/JC-201904-citibike-tripdata.csv")
# trip_2019_05 = pd.read_csv("Resources/JC-201905-citibike-tripdata.csv")
# trip_2019_06 = pd.read_csv("Resources/JC-201906-citibike-tripdata.csv")
```

```python
# Concatenate all csv files (Jan 2019 to June 2020)
all_qtr = pd.concat([trip_2019_01,trip_2019_02,trip_2019_03,trip_2019_04,trip_2019_05,trip_2019_06,trip_2019_07,trip_2019_08,trip_2019_09,
trip_2019_10,trip_2019_11,trip_2019_12,trip_2020_01,trip_2020_02,trip_2020_03,trip_2020_04,trip_2020_05,trip_2020_06])
```

```python
# Replace gender: 0 = Unknown; 1= Male; 2= Female
all_qtr.gender = all_qtr.gender.replace(0 ,'Unknown').replace(1 ,'Male').replace(2 ,'Female')
```

```python
# View last 20 records
all_qtr.tail(20)
#Ride_2019_04_06
```

```python
# Make a copy of histories df
history_df = all_qtr.copy()

history_df.head()
```

```python
# Format Start Time
history_df['Date'] = pd.to_datetime(history_df['starttime'])
history_df['Date'] = history_df['Date'].dt.date
history_df['Date'] = pd.to_datetime(history_df['Date'])
history_df['Date'] = history_df['Date'].dt.strftime('%m/%d/%Y')
```

```python
# Drop startime & stoptime column
history_df.drop(['starttime'], axis=1, inplace=True)
history_df.drop(['stoptime'], axis=1, inplace=True)


history_df
```

```python
x=history_df.groupby('Date')['bikeid'].nunique()
x
```

```python
Ride_2019_04_06['Date'] = pd.to_datetime(Ride_2019_04_06['Date'])
Ride_2019_04_06['Date'] = Ride_2019_04_06['Date'].dt.strftime('%m/%d/%Y')

Ride_2019_04_06.head()
```

```python
merged_df = Ride_2019_04_06.merge(history_df, on='Date')
merged_df
```

```python
# Export dataframe "all_qtr" to csv file
all_qtr.to_csv (r'Citi_Bike_Data.csv', index = False, header=True)
```

```python
# Export dataframe "merged_df" to csv file
merged_df.to_csv (r'Ride.csv', index = False, header=True)
```
